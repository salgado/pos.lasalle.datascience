### 1.2 - Hierarquia: Partes da Figura

Hierarquia: Cada gráfico é composto por 3 estruturas (objetos):

1. **Figure** - É a "caixa" principal que mantém o controle de todos os eixos filhos, dos "artistas" (títulos, legendas das figuras, etc.) e da tela. Uma figura irá conter uma ou mais plotagens.


2. **Axes** - Plotagens reais. A região da imagem com o espaço de dados. Um objeto **Figure** pode conter muitos **Axes**, mas um determinado objeto Axes só pode estar em um Figure. **Axes** contém dois (ou três no caso de 3D) objetos de **axis** (esteja ciente da diferença entre **Axes** e **Axis**) que cuidam dos limites de dados.


3. **Objetos menores** - "*tick marks*", "*individual lines*", "*legends*", and "*text boxes*".

### 1.3 - Diferença de Interfaces

Esta parte costuma gerar muita confusão quando estamos construindo gráficos mais elaborados utilizando a biblioteca matplotlib. Isto acontece pois a biblioteca é enorme, e muitas vezes temos formas diferentes de fazer a mesma coisa. 

<br>

De modo geral, temos duas formas de utilizar o matplotlib:

- **Object-oriented interface:** Criar explicitamente figuras e eixos, e chamar métodos para eles (De forma OO - orientado a objeto)

- **Pyplot Interface:** O Pyplot vai criar e gerenciar automaticamente as figuras e eixos e vamos utilizar as funções do pyplot para fazer a plotagem



**Exemplo para comparar os dois tipos de interface (Comparando O.O com Pyplot interface):**

Queremos construir dois gráficos, um do lado do outro. O primeiro gráfico é a função quadrática da variável x criada, e o segundo gráfico a função cúbica. 

```python
## Pyplot Interface - baseada em estágios

# Criando o objeto figure
plt.figure(figsize=(10, 4))          # figsize: tamanho da tela (largura, altura) em polegadas 

# Criando e editando o primeiro subplot:
plt.subplot(1,2,1)                          # (número de linhas, número de colunas, posição)
plt.plot(x, x**2, 'r-',label='Quadrado')    # Gráfico da função quadrada
plt.title('Primeiro Gráfico')
plt.legend()

# Criando e editando o segundo subplot:
plt.subplot(1,2,2)                          # (1 linha, 2 colunas, gráfico 2)
plt.plot(x, x**3, 'b-',label='Cubo')        # gráfico da função cúbica
plt.title('Segundo Gráfico')
plt.legend()
```

```python
## Object-oriented interface (Faz o formato geral inicialmente e vai modificado cada axes individualmente)

# Criando o objeto fig (figure) e axes (ax)
fig, ax = plt.subplots(nrows = 2, ncols = 1, figsize=(10,4))


# Editando o primeiro subplot:
ax[0].plot(x, x**2, 'r-',label='Quadrado')    # Gráfico da função quadrada
ax[0].set_title('Primeiro Plot')
ax[0].legend()


# # Editando o segundo subplot:
ax[1].plot(x, x**3, 'b-',label='Cubo')        # gráfico da função cúbica
ax[1].set_title('Segundo Plot')
ax[1].legend()


# Título geral
fig.suptitle('Gráficos feitos pela interface O.O', fontsize=14,y= 1.05,weight = 'bold', color = '#414040')

```

## <font color=green>  3 - Tipos de Gráficos

*   **Texto Simples**


> Usados quando queremos mostrar apenas um ou dois números, em destaque no texto.


*   **Heatmap**


> Uma forma de facilitar a visualização de tabelas, onde os numeros serão coloridos de acordo com a sua magnitude. 



*   **ScatterPlot**


> Ótimos para mostrar a relação entre duas variáveis. 




*   **Gráfico de Linha**


> Muito utilizados para plotar dados contínuos. Implicando conexão entre os pontos o que pode não fazer sentido para dados categóricos.  
Podem ser utilizados para mostrar apenas uma serie de dados ou múltiplas. Deve-se levar em consideração que o intervalo de tempo entre os pontos no eixo x, deve ser igualmente espaçado.

*   **Gráficos de barras**


> Talvez o tipo de gráfico mais utilizado e muitas vezes evitado por este motivo. Contudo são os gráficos de interpretação mais simples e que não exigem um grande esforço da audiência para entendê-los. Importante destacar que a escala do gráfico de barras **sempre** deve começar em zero.
Podem ser apresentados na forma de vertical, horizontal e empilhados. 


### Texto Simples




---


Imaginem que temos que prepara uma apresentação ou documento sobre as atividades offshore em um campo de petróleo e queremos destacar a seguinte frase: 

**"Digitalization can save up 80% of a project time"**

 
Neste caso, podemos usar um gráfico como um texto simples, para destacar esta informação. 


---








```python
# Importando as Bibliotecas 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import warnings 
warnings.filterwarnings("ignore")  # essa biblioteca impede que os avisos apareçam quando o código é rodado
```

```python
"""Criando o nosso objeto"""

# Quando você roda o sns.set(style...), as informações são atualizadas e todos os outros gráficos terão o formato. Para limpar a formatação, usamos o código abaixo 
sns.reset_orig()
sns.reset_defaults()

fig, ax = plt.subplots(figsize=(7, 5), nrows = 1, ncols= 1, dpi=110)
# figsize - tamanho da figura ((largura, altura) em polegadas) , dpi - resolução da imagem (pontos/polegada), quanto maior mais pesada e mais demora a rodar

# Removendo as bordas do gráfico
## Forma 1:
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# Forma 2:
# for spine in ax.spines.values():
#     spine.set_visible(False)

# Título do gráfico
ax.text(0.02, 0.85,
         'Offshore Activities         ',   # texto que queremos
         fontsize=26,     # tamanho do texto
         color='white',   # cor do texto
         # Colocando um retangulo de cor envolta do texto - para dar destaque ao título
         bbox={'facecolor': '#828282', 'pad': 10, 'edgecolor': 'none'})

# Adicionando nossa frase
ax.text(0.05, 0.65,
         'Digitalization can save up ',
         fontsize=19,
         color='#4A81BF')

# Adicionando nosso número que queremos destacar na apresentação - Negrito
ax.text(0.02, 0.30,
          '80%',
         fontsize=100,
         color='#4A81BF',weight='bold')

# Adicionando nossa frase
ax.text(0.05, 0.18,
         'of a project time.',
         fontsize=19,
         color='#4A81BF')

# Removendo os eixos do gráfico
ax.axis('off')

plt.savefig('Figura_texto.png')  # Salvando as figuras 
```

Observações:

- Estamos enfatizando o 80% para mostrar o impacto da digitalização no tempo de projeto. Para isto, utilizamos a estratégia de **aumentar o tamanho** e colocar em **negrito**. Assim, fazemos a audiência focar no número.

### Scatter Plot (Gráfico de dispersão)

---------
Será que existe relação entre a conta total dos clientes e a gorjeta dada em um determinado restaurante?

Para isto vamos analisar o dataset **tips** do pacote seaborn

- Descrição das variáveis:

  - Total_bill: Valor total da refeição (em dólares).

  - Tip: Valor da gorjeta (em dólares).

  - Sex: Sexo da pessoa pagando a refeição (Feminino / Masculino).

  - Smoker: Área para fumantes ou não.

  - Day: Dia da semana

  - Time: Almoço (Lunch) ou jantar (Dinner).

  - Size: Tamanho da refeição.

---------

```python
# Carregando o dataset
df1 = sns.load_dataset("tips")
df1.head()
```

```python
# Gráfico de Dispersão  (Scatterplot)

# ---------------------------------------------------------------------------------------
#                             Elementos básicos:
# ---------------------------------------------------------------------------------------

# 1 - Figure e axis
# Criando a variável fig e a variável ax   (figura, eixos):
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize=(5,3),       # (largura, altura) em polegadas 
                       dpi=110)                                   # resolução (pontos/polegada)

# 2 - Plot
# Gráfico de dispersão - total_bill (x), tip (y) de df1 
sns.scatterplot(data = df1, x="total_bill", y="tip", s = 45, linewidth = 0, color = '#A6A6A5' , ax = ax)    # cor opcional


# 3 - Título
# Inserindo título no nosso eixo ax
ax.set_title('Relação entre gorjeta e conta do restaurante', color = '#646369',loc='left', pad = 15,fontsize=13, weight='bold')


# ---------------------------------------------------------------------------------------
#                    Melhorando o aspecto do gráfico:
# ---------------------------------------------------------------------------------------

# Modificar a cor dos tiques para cinza (colocar uma cor mais clara - chamar menos atenção):
ax.tick_params(axis='both', colors = '#646369',labelsize=10)         # modifica a aparência dos ticks, ticks labels e gridlines


# Remoção de eixos desnecessários (Closure):
ax.spines['right'].set_visible(False)       # deixa invisível os eixos direito
ax.spines['top'].set_visible(False)         # deixa invisível os eixos superior

# Modificar a cor dos eixos inferior e esquerdo para cinza
ax.spines['bottom'].set_color('#646369')       # inferior
ax.spines['left'].set_color('#646369')         # esquerdo

# Modificar o label
ax.set_xlabel('Valor da conta',color='#646369',fontsize=10,
              position=(0, 0), horizontalalignment='left')
ax.set_ylabel('Gorjeta', color='#646369', fontsize=10,
              position=(0,1), horizontalalignment='right')


# Formatar os ticks label (eixo x e y) para ficar $ com o valor

## Função geral a ser aplicada
def formatar_eixo(value, tick_number):      # (valor, posição)
    return ('$'+str(int(value)))            # para cada valor do tick_number, ele vai fazer '$tick_number'

## Aplicar a função formatar_eixo para formatar os ticks no xaxis e yaxis
ax.xaxis.set_major_formatter(plt.FuncFormatter(formatar_eixo))    # eixo x
ax.yaxis.set_major_formatter(plt.FuncFormatter(formatar_eixo))    # eixo y

# ---------------------------------------------------------------------------------------
#                    Informações adicionais que queremos passar:
# ---------------------------------------------------------------------------------------

# Qual a média do valor de gorjeta dada?
# Quais valores estão acima da média?

# Qual a média do valor de gorjeta dada?
media_tip = df1['tip'].mean()
media_total = df1['total_bill'].mean()

# Adicionando linha para indicar a média de gorjeta dada
ax.axhline(y = media_tip, color='#555655', linestyle='--',linewidth = 1)

# Gráfico para traçar o valor médio de (Valor da conta, gorjeta):
sns.scatterplot(x = [media_total], y = [media_tip], s = 90, linewidth=0, color = '#231F20' , ax = ax)

# Texto para indicar que é a média:
ax.text(media_total - 0.15 * media_total, media_tip + 0.2* media_tip,'Média',color='black', fontsize=11)
```

Vamos ver a diferença do antes e depois? 

- Você mudaria alguma coisa? A partir do que foi ensinado, faça suas mudanças. Tente pensar no que você quer enfatizar no gráfico. Quem sabe colocar um textinho indicando que existe relação entre o valor da conta e a gorjeta? 

```python
# Comparação antes e depois


# Figure e axis

# Criando 2 gráficos
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize=(11,3),       # (largura, altura) em polegadas 
                       dpi=110)                                   # resolução (pontos/polegada)

# ------------------------------------------------------------------------------------
#     Gráfico 1 - Simples

# Gráfico de dispersão - total_bill (x), tip (y) de df1: 
sns.scatterplot(data = df1, x="total_bill", y="tip", ax = ax[0])

# Título
ax[0].set_title('Antes')


# ------------------------------------------------------------------------------------
#    Gráfico 2 - Com modificações

# Gráfico de dispersão - total_bill (x), tip (y) de df1 
sns.scatterplot(data = df1, x="total_bill", y="tip", s=45, linewidth=0,color = '#A6A6A5', ax = ax[1])

# Título
ax[1].set_title('Depois', color = '#646369',loc='left', pad = 15, fontsize=13, weight='bold')

# modifica a aparência dos ticks, ticks labels e gridlines 
ax[1].tick_params(axis='both', colors ='#646369', labelsize=10)         

# Remoção de eixos desnecessários (Closure):
ax[1].spines['right'].set_visible(False)       # deixa invisível os eixos direito
ax[1].spines['top'].set_visible(False)         # deixa invisível os eixos superior

# Modificar a cor dos eixos inferior e esquerdo para cinza
ax[1].spines['bottom'].set_color('#646369')       # inferior
ax[1].spines['left'].set_color('#646369')         # esquerdo


# Modificar o label
ax[1].set_xlabel('Valor da conta',color='#646369',fontsize=10,position=(0, 0), horizontalalignment='left')
ax[1].set_ylabel('Gorjeta',color='#646369',fontsize=10, position=(0,1), horizontalalignment='right')

# Formatar os ticks label (eixo x e y) para ficar $ com o valor
def formatar_eixo(value, tick_number):
    return ('$'+str(int(value)))   # para cada valor do tick_number, ele vai fazer '$tick_number'

ax[1].xaxis.set_major_formatter(plt.FuncFormatter(formatar_eixo))    # eixo x
ax[1].yaxis.set_major_formatter(plt.FuncFormatter(formatar_eixo))    # eixo y

# Adicionando linha para indicar a média de gorjeta dada
ax[1].axhline(y = media_tip, color='#555655', linestyle='--',linewidth = 1)

# Gráfico para traçar o valor médio de (Valor da conta, gorjeta):
sns.scatterplot(x = [media_total], y = [media_tip], s = 90, linewidth=0, color = '#231F20' , ax = ax[1])

# Texto para indicar que é a média:
ax[1].text(media_total - 0.15 * media_total, media_tip + 0.2* media_tip,'Média',color='black', fontsize=11)
```

#### Gráfico de Barras (horizontais)



---
Utilizando o gráfico de barras horizontal e dispondo os dados em ordem decrescente pode facilitar a leitura dos dados. 

Além disso neste caso, queremos destacar que os setores de Beleza e Infantil são os que mais vendem

---





```python
# ordem decrescente:
df3 = df3.sort_values(2017, ascending=False)    # do maior para menor, em relação a 2017
df3
```

```python
# V3 - Fazendo mudanças para melhorar o gráfico


#criando uma figure, axes
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize=(5,4),       # (largura, altura) em polegadas 
                       dpi=110)                                   # resolução (pontos/polegada)


# criando o gráfico de barras (horizontal)

## Construindo nossa própria paleta de cores:
first_color = '#4A81BF'
other_colors = '#BFBEBE'
palette1 = 2*[first_color] + (len(df3)-2)*[other_colors]   # as duas primeiras cores serão 'first_color' e as demais 'other_colors'

# Gráfico de barras (notar que é o mesmo processo que o anterior, só vai inverter o x e y)
sns.barplot(x = df3[2017], y = df3.index, ax=ax, palette = palette1,data = df3,order = df3.sort_values(2017,ascending = False).index)


# Remover linhas

# Modificar os eixos (linhas)

## Deixar invisível os eixos superior e direito
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

## Modificar a cor dos eixos inferior e esquerdo para cinza
ax.spines['bottom'].set_color('#cfd8dc')
ax.spines['left'].set_color('#cfd8dc')


# --------------- REPETINDO O PROCESSO ------------------------------
#Título -> notar que o título não está alinhado com o eixo

# Alinhar o texto
ax.text(x= -350,y = -1.2, s = 'Beleza e Infantil são os setores com maior quantidade\nvendida em 2017',color = '#555655',fontsize=14)

# Modificar a cor dos tiques para cinza (colocar uma cor menos intensa)
ax.tick_params(axis='both', colors ='#76787B',labelsize=10)         # modifica a aparência dos ticks, ticks labels e gridlines 


# Adicionar o label
ax.set_xlabel('',color='#76787B')
ax.set_ylabel('Setor',color='#76787B',fontsize=12)


# # ---------------------------------------------------------------------------------------------------
# #    Parte 2: Vamos escrever os números na barra (princípios da proximidade, continuidade, visão-Z)
# # ---------------------------------------------------------------------------------------------------

for i in range(len(df3)):   # posição i
  if i in [0,1]:            # colocar em negrito os setores beleza e infantil
    ax.text(df3[2017][i] - 10,i, df3[2017][i] , fontsize=11, color='white',verticalalignment='center',horizontalalignment='right',weight='bold')  
  else:                     # os demais são normais
    ax.text(df3[2017][i] - 10,i, df3[2017][i] , fontsize=11, color='white',verticalalignment='center',horizontalalignment='right')  


# Escrever quantidade para reforçar o número
ax.text(0,0-0.5, 'Quantidade', fontsize=11, color='#76787B')  


# Agora podemos tirar o eixo-x já que os valores estão todos nas barras:

ax.spines['bottom'].set_visible(False)

ax.tick_params(
    axis='x',            # modificar o eixo x
    which='both',        # modificar major and minor ticks 
    bottom=False,        # remover o tick parte inferior
    labelbottom=False,   # remover os números
    ) 
```

Obs: Podemos notar que conseguimos ler melhor na forma horizontal. Estudos de design mostram que normalmente se tem um padrão comum que os olhos percorrem as imagens (forma “Z”)

topo/esquerda ----> topo/direita---> base/esquerda ---> base/direita
