# Customizando a vizualização

```python
brasil
```

```python
fig, ax = plt.subplots(figsize=[6,4], dpi=100)
ax.plot(brasil['Year'], brasil['gdp'])
ax.set_xlabel("Ano", fontdict={'size':12})
ax.set_ylabel("GDP", fontdict={'size':12})
ax.set_title("GDP (Brasil)")
plt.show()
```

```python
# marcadores : https://matplotlib.org/stable/api/markers_api.html

fig, ax = plt.subplots(figsize=[6,4], dpi=100)
ax.plot(brasil['Year'], brasil['gdp'], marker ='o')
ax.set_xlabel("Ano", fontdict={'size':12})
ax.set_ylabel("GDP", fontdict={'size':12})
ax.set_title("GDP (Brasil)")
plt.show()
```

```python
# linhas tracejadas
# e coloridas

fig, ax = plt.subplots(figsize=[6,4], dpi=100)
ax.plot(brasil['Year'], brasil['gdp'], marker ='o', linestyle="--", color="g")
ax.set_xlabel("Ano", fontdict={'size':12})
ax.set_ylabel("GDP", fontdict={'size':12})
ax.set_title("GDP (Brasil)")
plt.show()
```

```python
# grid
# ticks

fig, ax = plt.subplots(figsize=[6,4], dpi=100)
ax.plot(brasil['Year'], brasil['gdp'], marker ='o', linestyle="--", color="g")
ax.set_xticks(list(range(brasil['Year'].min(), brasil['Year'].max()+1, 6)))
ax.set_xlabel("Ano", fontdict={'size':12})
ax.set_ylabel("GDP", fontdict={'size':12})
ax.set_title("GDP (Brasil) / Per Capita")
plt.grid("True")
plt.show()
```

```python
# labels, legenda

fig, ax = plt.subplots(figsize=[6,4], dpi=100)
ax.plot(brasil['Year'], brasil['gdp'], marker ='o', linestyle="--", color="g", label='Brasil')
ax.set_xticks(list(range(brasil['Year'].min(), brasil['Year'].max()+1, 6)))
ax.set_xlabel("Ano", fontdict={'size':12})
ax.set_ylabel("GDP", fontdict={'size':12})
ax.set_title("GDP (Brasil) / Per Capita")
ax.legend(loc='best')
plt.grid("True")
plt.show()
```

```python
# labels, legenda

fig, ax = plt.subplots(figsize=[6,4], dpi=100)
ax.plot(brasil['Year'], brasil['gdp'], marker =None, linestyle="--", color="g", label='Brasil')
ax.set_xticks(list(range(brasil['Year'].min(), brasil['Year'].max()+1, 6)))
ax.set_xlabel("Ano", fontdict={'size':12})
ax.set_ylabel("GDP", fontdict={'size':12})
ax.set_title("GDP (Brasil) / Per Capita")
#ax.legend(loc='best')
handles, labels= ax.get_legend_handles_labels()
ax.legend(handles, labels, title="País", title_fontsize="medium") #small, medium, large
plt.grid("True")
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
# %matplotlib notebook

import pandas as pd
```

```python
arq="/content/drive/MyDrive/Aulas/pos/datascience/dataviz/data/gapminder.csv"
df = pd.read_csv(arq)
brasil=df.query("Country =='Brazil' ")
```

```python
df_brazil = df.query("Country =='Brazil' ").reset_index(drop=True)
df_brazil
```

```python
df_usa = df.query("Country =='United States' ").reset_index(drop=True)
df_usa
```

```python
fig, ax = plt.subplots(figsize=[6,4], dpi=100)
ax.plot(df_brazil['Year'], df_brazil['gdp'], marker =None, linestyle="-", color="b", label='Brasil')
ax.plot(df_usa['Year'], df_usa['gdp'], marker =None, linestyle="-", color="r", label='USA')
ax.set_xlabel('Ano')
ax.set_ylabel('GDP Per Capita')
ax.legend(title="País")
plt.show()
```

```python
#separando plots
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10,8))
ax[0].plot(df_brazil['Year'], df_brazil['gdp'], marker =None, linestyle="-", color="b", label='Brasil')
ax[1].plot(df_usa['Year'], df_usa['gdp'], marker =None, linestyle="-", color="r", label='USA')
ax[1].set_xlabel('Ano')
ax[0].set_ylabel('GDP Per Capita Brasil')
ax[1].set_ylabel('GDP Per Capita EUA')
ax[0].set_title ("Evlução do GDP Per Capita / Brasil x EUA")
#ax[0].legend(title="País")
plt.show()
```

```python
#separando plots
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10,8))
ax[0].plot(df_brazil['Year'], df_brazil['gdp'], marker =None, linestyle="-", color="b", label='Brasil')
ax[1].plot(df_usa['Year'], df_usa['gdp'], marker =None, linestyle="-", color="r", label='USA')
ax[1].set_xlabel('Ano')
ax[0].set_ylabel('GDP Per Capita Brasil')
ax[1].set_ylabel('GDP Per Capita EUA')
ax[0].set_title ("Evlução do GDP Per Capita / Brasil x EUA")
ax[0].grid(True)
ax[1].grid(True)

plt.show()
```
