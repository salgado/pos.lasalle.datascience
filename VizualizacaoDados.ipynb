```python
fig, ax = plt.subplots(figsize=[12,8], dpi=80)
ax.plot(df_brazil['Year'], df_brazil['gdp'])
plt.show()
```

```python
print(plt.rcParams.get('figure.figsize'))
```

```python
# adicionar labels
fig, ax = plt.subplots(figsize=[12,8], dpi=70)
ax.plot(df_brazil['Year'], df_brazil['gdp'])

ax.set_xlabel('Ano', fontdict={'size': 12})
ax.set_ylabel('PIB', fontdict={'size': 12})
ax.set_title('PIB Brasil Evolução (1964-2013)')
plt.show()
```

```python
# marcadores
fig, ax = plt.subplots(figsize=[12,8], dpi=70)
ax.plot(df_brazil['Year'], df_brazil['gdp'], marker='*')

ax.set_xlabel('Ano', fontdict={'size': 12})
ax.set_ylabel('PIB', fontdict={'size': 12})
ax.set_title('PIB Brasil Evolução (1964-2013)')
plt.show()
```

```python
# marcadores
fig, ax = plt.subplots(figsize=[12,8], dpi=70)
ax.plot(df_brazil['Year'], df_brazil['gdp'], linestyle="--", color="g")

ax.set_xlabel('Ano', fontdict={'size': 12})
ax.set_ylabel('PIB', fontdict={'size': 12})
ax.set_title('PIB Brasil Evolução (1964-2013)')
plt.show()
```

```python
# grid, ticks
fig, ax = plt.subplots(figsize=[12,8], dpi=70)
ax.plot(df_brazil['Year'], df_brazil['gdp'], linestyle="--", color="g")

ax.set_xticks(list(range(df_brazil['Year'].min(), df_brazil['Year'].max()+1, 6  )))
ax.set_xlabel('Ano', fontdict={'size': 12})
ax.set_ylabel('PIB', fontdict={'size': 12})
ax.set_title('PIB Brasil Evolução (1964-2013)')
plt.grid(True)
plt.show()
```

```python
# legenda, label

fig, ax = plt.subplots(figsize=[12,8], dpi=70)
ax.plot(df_brazil['Year'], df_brazil['gdp'], linestyle="-", color="g", label='Brasil')

ax.set_xticks(list(range(df_brazil['Year'].min(), df_brazil['Year'].max()+1, 6  )))
ax.set_xlabel('Ano', fontdict={'size': 12})
ax.set_ylabel('PIB', fontdict={'size': 12})
ax.set_title('PIB Brasil Evolução (1964-2013)')
#ax.legend(loc='best')
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels, title='País', title_fontsize='medium') # small, medium, large

plt.grid(True)
plt.show()
```

# Múltiplos plots

```python
df_usa = df.query(" Country == 'United States' ").reset_index(drop=True)
df_usa
```

```python
#plotar brasil e usa

fig, ax = plt.subplots(figsize=[12,8], dpi=70)
ax.plot(df_brazil['Year'], df_brazil['gdp'], linestyle="-", color="g", label='Brasil')
ax.plot(df_usa['Year'], df_usa['gdp'], linestyle="-", color="r", label="EUA")


ax.set_xticks(list(range(df_brazil['Year'].min(), df_brazil['Year'].max()+1, 6  )))
ax.set_xlabel('Ano', fontdict={'size': 12})
ax.set_ylabel('PIB', fontdict={'size': 12})
ax.set_title('PIB Evolução (BrasilxEUA) (1964-2013)')
#ax.legend(loc='best')
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels, title='País', title_fontsize='medium') # small, medium, large

plt.grid(True)
plt.show()
```

```python
#dividir plot

fig, ax = plt.subplots( nrows=2, ncols=1, figsize=[12,8], dpi=70)
ax[0].plot(df_brazil['Year'], df_brazil['gdp'], linestyle="-", color="g", label='Brasil')
ax[1].plot(df_usa['Year'], df_usa['gdp'], linestyle="-", color="r", label="EUA")

# ax.set_xticks(list(range(df_brazil['Year'].min(), df_brazil['Year'].max()+1, 6  )))
ax[1].set_xlabel('Ano', fontdict={'size': 12})
ax[0].set_ylabel('PIB Brasil', fontdict={'size': 12})
ax[1].set_ylabel('PIB EUA', fontdict={'size': 12})
ax[0].set_title('PIB Evolução (BrasilxEUA) (1964-2013)')
#ax.legend(loc='best')
#handles, labels = ax.get_legend_handles_labels()
#ax.legend(handles, labels, title='País', title_fontsize='medium') # small, medium, large

ax[0].grid(True)
ax[1].grid(True)
#plt.grid(True)
plt.show()
```

```python
#twinx

fig, ax = plt.subplots( figsize=[12,8], dpi=70)
ax.plot(df_brazil['Year'], df_brazil['gdp'], linestyle="-", color="g", label='Brasil')
ax.set_xlabel('Ano', fontdict={'size': 12})
ax.set_ylabel('PIB Brasil', fontdict={'size': 12})
ax.set_title('PIB Evolução (BrasilxEUA) (1964-2013)')

ax2 = ax.twinx()

ax2.plot(df_usa['Year'], df_usa['gdp'], linestyle="-", color="r", label="EUA")
ax2.set_ylabel('PIB EUA', fontdict={'size': 12})

handles, labels = ax.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()

ax.legend(handles + handles2, labels + labels2, loc='best', title='País', title_fontsize='medium') # small, medium, large

plt.grid(True)
plt.show()
```
